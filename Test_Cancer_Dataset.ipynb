```python
import os
import json
import numpy as np

# Paths to the folders
npz_folder = '/raid/mpsych/OMAMA/DATA/data/2d_resized_512/images'
json_folder = '/raid/mpsych/OMAMA/DATA/data/2d/metadata/'

# Function to read JSON files
def read_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Lists to store filenames for cancer and non-cancer cases
cancer_files = []
non_cancer_files = []

# Function to classify NPZ files
def classify_npz_files():
    # Read JSON files
    json_files = os.listdir(json_folder)
    files_loaded = 0
    for json_file in json_files:
        if not json_file.endswith('.json'):
            continue
        json_data = read_json(os.path.join(json_folder, json_file))
        
        # Extract label and filename
        label = json_data['label']
        filename = json_file.replace('.json', '.npz')
        
        # Match with NPZ files
        npz_file_path = os.path.join(npz_folder, filename)
        if os.path.exists(npz_file_path):
            # Classify based on label
            if label == 'NonCancer':
                if len(non_cancer_files) < 10:
                    non_cancer_files.append(npz_file_path)
                    files_loaded += 1
            else:
                if len(cancer_files) < 10:
                    cancer_files.append(npz_file_path)
                    files_loaded += 1

            if files_loaded >= 20:
                break
        else:
            print(f"NPZ file corresponding to {json_file} not found.")

# Classify NPZ files
classify_npz_files()

# Display the loaded filenames for cancer and non-cancer cases
print("Cancer Files:")
print(cancer_files)
print("\nNon-Cancer Files:")
print(non_cancer_files)
```

```python
import os
import json
import numpy as np

# Paths to the folders
npz_folder = '/raid/mpsych/OMAMA/DATA/data/2d_resized_512/images'
json_folder = '/raid/mpsych/OMAMA/DATA/data/2d/metadata/'

# Function to read JSON files
def read_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Lists to store filenames for cancer, non-cancer, and unknown cases
cancer_files = []
non_cancer_files = []
unknown_files = []

# Function to classify NPZ files
def classify_npz_files():
    # Read JSON files
    json_files = os.listdir(json_folder)
    for json_file in json_files:
        if not json_file.endswith('.json'):
            continue
        json_data = read_json(os.path.join(json_folder, json_file))
        
        # Extract label and filename
        label = json_data['label']
        filename = json_file.replace('.json', '.npz')
        
        # Match with NPZ files
        npz_file_path = os.path.join(npz_folder, filename)
        if os.path.exists(npz_file_path):
            # Classify based on label
            if label == 'NonCancer':
                non_cancer_files.append(npz_file_path)
            elif label == 'Unknown':
                unknown_files.append(npz_file_path)
            else:
                cancer_files.append(npz_file_path)                 
        else:
            print(f"NPZ file corresponding to {json_file} not found.")

# Classify NPZ files
classify_npz_files()

# Print the number of total files loaded
total_files_loaded = len(cancer_files) + len(non_cancer_files) + len(unknown_files)
print("Total files loaded:", total_files_loaded)
```
